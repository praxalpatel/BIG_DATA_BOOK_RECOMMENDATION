```python
def get_list(ls):
    temp=[]
    temp.append(int(ls[0][2:]))
    temp.append(float(ls[1][:-1]))
    temp.append(float(ls[3]))
    temp.append(float(ls[5]))
    #print(ls[7][:-2])
    temp.append(float(ls[7][:-2]))
    return temp
    
```

```python
data=pd.DataFrame(columns=['rank','reg','map_score','ndcg_score','mpa_score'])
with open("results.txt") as file:
    for line in file:
        print(line)
        ls=line.split(',')
        temp=pd.DataFrame([get_list(ls)],columns=data.columns)
        data=data.append(temp,ignore_index=True)
data
```

```python
data['rank'].value_counts()
```

```python
def map_trend_1():
    nc_comp=data['reg'].unique()
    for nc in nc_comp:
        temp="reg: "+str(nc)
        logistic=data[(data['reg']==nc)]
        plt.scatter(logistic['rank'],logistic['ndcg_score'],label=temp,marker='o')
        plt.legend(loc='best')
        plt.xlabel("Rank")
        plt.ylim([0,0.03])
        plt.ylabel("ndcg Score")
        plt.title("NDCG vs rank for varying value of regularization")
        plt.savefig('ndcg_scores.png')
map_trend_1()
```

### Lightfm results for 0.1 percent data

```python
df=pd.read_csv("lightfm_results_point1.csv")
df.head()
```

### AUC curve for lightfm loss functions for 5 components

```python
def plot_lightfm(nc):
    warp=df[(df['loss']=='warp') & (df['components']==nc)]
    bpr=df[(df['loss']=='bpr') & (df['components']==nc)]
    logistic=df[(df['loss']=='logistic') & (df['components']==nc)]
    plt.plot(warp['epochs'],warp['auc'],label='warp_auc')
    plt.plot(bpr['epochs'],bpr['auc'],label='bpr_auc')
    plt.plot(logistic['epochs'],logistic['auc'],label='logistic_auc')
    plt.legend(loc='best')
plot_lightfm(5)
```

### AUC curve for lightfm loss functions for 10 components

```python
plot_lightfm(10)
```

### AUC curve for lightfm logitstic loss trend 

```python
def nc_trend(loss):
    nc_comp=df['components'].unique()
    for nc in nc_comp:
        temp="num_componenets: "+str(nc)
        logistic=df[(df['loss']==loss) & (df['components']==nc)]
        plt.plot(logistic['epochs'],logistic['auc'],label=temp,marker='o')
        plt.legend(loc='best')
        plt.xlabel("Number of epochs")
        plt.ylabel("AUC Score")
        plt.title("AUC vs number of epochs using "+loss+" loss function for varying number of components")
```

```python
nc_trend('logistic')
```

### AUC curve for lightfm BPR loss trend 

```python
nc_trend('bpr')
```

### AUC curve for lightfm WARP loss trend 

```python
nc_trend('warp')
```

### AUC and time trend for lightfm warp loss function

```python
def auc_time_trend_new(loss,df):
    nc_comp=df['components'].unique()
    for nc in nc_comp:
        temp="num_componenets: "+str(nc)
        logistic=df[(df['loss']==loss) & (df['components']==nc)]
        plt.plot(logistic['time'],logistic['auc'],label=temp,marker='o')
        plt.legend(loc='best')
        plt.xlabel("Time")
        plt.ylabel("AUC Score")
        plt.title("AUC vs number of epochs using "+loss+" loss function for varying number of components")
```

```python
auc_time_trend_new('warp',df)
```
